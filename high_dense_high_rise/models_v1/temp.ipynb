```python
import os
import pandas as pd
```

```python
cur_dir =os.getcwd()
model_dfs = []
for file in os.listdir(cur_dir):
    if file.endswith('.pth'):
        model_name = file.replace('.pth','')
        _, weight, epochs, acc, prec, recall = model_name.split('_')
        model_dfs.append({
            'Model': model_name+'.pth',
            'weight': float(weight),
            'epoch': int(epochs.replace('epoch', '')),
            'acc': float(acc.replace('acc', '')),
            'precision': float(prec.replace('prec', '')),
            'recall': float(recall.replace('rec', ''))
        })
model_dfs = pd.DataFrame(model_dfs)
model_dfs
```

```python
model_dfs.to_csv('model_df.csv', index=False)
```
